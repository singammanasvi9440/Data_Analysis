```python
import pandas as pd
df=pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df.drop_duplicates()
df.drop(columns='Not_Useful_Column',inplace=True)
df.replace('Y','Yes',inplace=True)
df.replace('N','No',inplace=True)
```

```python

df
```

```python
# #df['Last_Name'].replace('/','')
# df['Last_Name']=df['Last_Name'].str.lstrip('...')
# df['Last_Name']=df['Last_Name'].str.lstrip('/')
# df['Last_Name']=df['Last_Name'].str.rstrip('...')
# df['Last_Name']=df['Last_Name'].str.rstrip('_')
# # df
df['Last_Name']=df['Last_Name'].str.strip('12./_')
```

```python
 df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')\

 df
```

```python

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df
```

```python
df["Address"] = df["Address"].replace('N/a','')
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','')
df=df.fillna('')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)

df
```

```python
df.drop(columns='Address',inplace=True)
df
```
