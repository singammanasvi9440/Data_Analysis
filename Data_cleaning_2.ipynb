```python
import pandas as pd
df= pd.read_csv("/content/unclean_data.csv",encoding='latin1')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns={'DIRECTOR_facebook_likes':'DIRECTOR_FB_LIKES','no.of.voted_users':'no_of_voted_users'},inplace=True)
df.rename(columns={'actor_3_facebook_likes':'Actor_3_FB_Likes','ACTOR_1_facebook_likes':'ACTOR_1_FB_likes'},inplace=True)

df.columns
```

```python
df.drop(columns='title_year.1',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.fillna(0, inplace = True)
```

```python
df['duration'].mean()
```

```python
df.fillna(df['duration'].mean())
```

```python
df.drop_duplicates('movie_title',inplace=True)
```

```python
df['movie_title']=df['movie_title'].str.replace('[^a-zA-Z0-9]','')
df = df.astype(str)
```

```python
df.shape
```

```python
df.dtypes
```

```python
col=['num_voted_users','Cast_Total_facebook_likes','facenumber_in_poster','ACTOR_2_facebook_likes']
```

```python
df[col] = df[col].astype(int) # modifying data type of multiple columns at a time  on original dataset
```

```python
df.dtypes
```

```python
# modifying data type of each columns
df.num_voted_users.astype(int)
df.Cast_Total_facebook_likes.astype(int)
df.facenumber_in_poster.astype(int)
df.ACTOR_2_facebook_likes.astype(int)
```
