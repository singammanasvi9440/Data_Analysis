```python
import pandas as pd
df= pd.read_csv("/content/unclean_data.csv",encoding='latin1')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns={'DIRECTOR_facebook_likes':'DIRECTOR_FB_LIKES','no.of.voted_users':'no_of_voted_users'},inplace=True)
df.rename(columns={'actor_3_facebook_likes':'Actor_3_FB_Likes','ACTOR_1_facebook_likes':'ACTOR_1_FB_likes'},inplace=True)

df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns.inplace = True
```

```python
df.drop(columns='title_year.1',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.fillna(0, inplace = True)
```

```python
df['duration'].mean()
```

```python
df.fillna(df['duration'].mean())
```

```python
df.drop_duplicates('movie_title',inplace=True)
```

```python
df['movie_title'] = df['movie_title'].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
df.dtypes
```

```python
col=['num_voted_users','cast_total_facebook_likes','facenumber_in_poster','actor_2_facebook_likes']
```

```python
df[col] = df[col].astype(int) # modifying data type of multiple columns at a time  on original dataset
```

```python
df.dtypes
```

```python
df
```

```python
# modifying data type of each columns
df.num_voted_users.astype(int)
df.cast_total_facebook_likes.astype(int)
df.facenumber_in_poster.astype(int)
df.actor_2_facebook_likes.astype(int)
```

```python
df.hist('imdb_score')
```

```python
import matplotlib.pyplot as plt
df.plot('duration')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.bar(df['movie_title'], df['duration'], color='skyblue')
plt.xlabel('Movie Name')
plt.ylabel('Duration (minutes)')
plt.title('Movie Duration')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df['movie_title'], df['actor_3_fb_likes'], label='Actor_3_FB_Likes', marker='o')
plt.plot(df['movie_title'], df['actor_1_fb_likes'], label='ACTOR_1_FB_likes', marker='o')
plt.plot(df['movie_title'], df['actor_2_facebook_likes'], label='ACTOR_2_facebook_likes', marker='o')

# Add labels and title
plt.xlabel('Movie ID')
plt.ylabel('Facebook Likes')
plt.title('Comparision of Facebook Likes')
plt.xticks(rotation=90, ha='right')
plt.legend()

plt.grid(True)  # Add grid lines for better readability
plt.show()
```

```python
plt.scatter(df['budget'],df['movie_title'])
plt.show()
```

```python
# Group movies by release year
grouped_df = df.groupby('title_year')['movie_title'].apply(list).reset_index(name='Movies')
print(grouped_df)
```

```python
plt.figure(figsize=(10, 6))
bar_width = 0.35

# Plot Budget
plt.bar(df['movie_title'], df['budget'], width=bar_width, label='Budget', color='blue')

# Plot Gross
plt.bar([i + bar_width for i in range(len(df))], df['gross'], width=bar_width, label='Gross', color='orange')

# Customize the plot
plt.xlabel('Movie')
plt.ylabel('Amount (in millions)')
plt.title('Budget and Gross for Each Movie')

plt.xticks(rotation=90, ha='right')
df.sort_values(by='gross', ascending=False)
plt.legend()

# Show the plot
plt.show()
```
