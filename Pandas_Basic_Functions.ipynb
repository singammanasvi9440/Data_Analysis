Write a Pandas program to create and display a DataFrame from a specified dictionary data which has the index labels.
Sample Python dictionary data and list labels:
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']



```python
import numpy as np
Data={
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(Data , index=labels)
print(df)
```

```python
df.info()   #summary of DF
```

```python
df.iloc[:3]  #get the first 3 rows of a given DataFrame.
```

```python
df[['name','score']] #select the 'name' and 'score' columns from the DataFrame
```

```python
#Select 'qualify' and 'score' columns in rows 1, 3, 5, 6.
df.iloc[[1, 3, 5, 6], [1, 3]]     #select the specified columns and rows from a given data frame
```

```python
#select the rows where the number of attempts in the examination is greater than 2.

df[df['attempts'] > 2]
```

```python
#score is between 15 and 20 (inclusive)
df[df['score'].between(15, 20)]
```

```python
#score is between 15 and 20 (inclusive)
df[df['score']>15 & (df['attempts']<2)]
```

```python
#number of rows and columns of a DataFrame
df.shape
```

```python
df[df['score'].isnull()] #Select the rows where the score is missing, i.e. is NaN.
```

```python
#\nChange the score in row 'd' to 11.5:
df.loc['d', 'score'] = 11.5
df
```
