

**Replacing Nan values uisng different methods**

```python
new_df=df.fillna(0)  #replacing Nan values with 0
print("Nan values Dataframe\n",new_df)
```

```python
newdf1= df.fillna({        #replacing Nan values in temperature & windspeed with 0 and event with No event.
 'temperature': 0,
     'windspeed': 0,
      'event':'no event'
    })
newdf1
```

```python
#Basically if we fill the Nan values with 0 it doesn't make any sense.For ex,let's say temp on 2017-01-01 is 32.0.
#But on 2017-01-04 is 0 degree which wouldn't happen in real world.
#So,lets forward fill the temperature using a method "Forwardfill->ffill" which takes previous days value to the next.
#This mehtods copies values vertically by default.

newdf2= df.fillna(method='ffill')
print("Forward_fill: \n", newdf2)

#similarly we can fill using "backwardfill->bfill"
newdf3= df.fillna(method='bfill')
print("Backward_fill: \n",newdf2)

#If you want to copy horizantally, use axis=columns,this way is mostly not used.
newdf4= df.fillna(method='ffill',axis='columns')
print("Forward_fill horizontally : \n", newdf4)



```

```python
#interpolation is a technique used to estimate missing or undefined values in a dataset.
#By default it performs linear interpolation.
#In general, interpolation methods work best when the data is sorted.Here the data is sorted(day)
#When the data is not sorted,the assumptions may lead to inaccurate results.
newdf5=df.interpolate()
print("Interpolated data: \n",newdf5)
```

```python
#using time method with interploate gives accurate results as here the dates
# are not continuous.
newdf5=df.interpolate(method="time")
print("Interpolated data: \n",newdf5)
```
